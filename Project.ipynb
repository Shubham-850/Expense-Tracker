```python
from faker import Faker
import random
```

```python
import pandas as pd
```

```python
#initialize faker
fake = Faker()
```

```python
categories = ["Food", "Transportation", "Bills", "Entertainment", "Groceries", "Subscriptions","investment"]
payment_modes = ["Cash", "Credit Card", "Debit Card", "UPI","Net Banking"]
cashback_rates = [10,12.5,11.4,10.4,3,7,5,8,3,5.3,6.3,7.34,6.3,6.4,.6,1.7,1.6,12.67,2.7,2,6.3,]
```

```python
expense = {
    "Date" : fake.date_this_month,
    "Category" : random.choice(categories),
    "Payment Mode" : random.choice(payment_modes),
    "Description" : fake.sentence(),
    "Amount" : round(random.uniform(10,1000),2),
    "Cashback" : random.choice(cashback_rates)
    }
print(expense)
```

```python
import datetime
```

```python
def gen_exp_data(num_entries = 150):
    start_date = datetime.date(2024, 12, 1)
    end_date = datetime.date(2024, 12, 31)
    data = []
    for _ in range(num_entries):
        expense = {
            "Date" : fake.date_between(start_date=start_date, end_date=end_date),
            "Category" : random.choice(categories),
            "Payment Mode" : random.choice(payment_modes),
            "Description" : fake.sentence(),
            "Amount" : round(random.uniform(10,1000),2),
            "Cashback" : random.choice(cashback_rates)
            }
        data.append(expense)
    return pd.DataFrame(data)
exp_data = gen_exp_data(150)
```
